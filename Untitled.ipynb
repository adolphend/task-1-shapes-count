```python
import pandas as pd
import seaborn as sn
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import numpy as np


def plot_confusion_matrix(conf_mat, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    This function plots the confusion matrix.

    Args:
        - conf_mat (np.array): matrix

        - classes (int): number of classes

        - normalize (bool): to apply a normalization on cm

        - title (str): title of the plot

        - cmap (plt.cm): to specify a plot cmap

    Returns:
        - axe (plt.axes): Axes object

    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    if normalize:
        conf_mat = conf_mat.astype(
            'float') / conf_mat.sum(axis=1)[:, np.newaxis]

    fig, axe = plt.subplots()
    img = axe.imshow(conf_mat, interpolation='nearest', cmap=cmap)
    axe.figure.colorbar(img, ax=axe)
    axe.set(xticks=np.arange(conf_mat.shape[1]),
            yticks=np.arange(conf_mat.shape[0]),
            xticklabels=classes, yticklabels=classes,
            title=title,
            ylabel='True label',
            xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(axe.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    plot_cm = np.round(conf_mat, decimals=2)
    fmt = 'float' if normalize else 'int'
    thresh = conf_mat.max() / 2.
    for i in range(conf_mat.shape[0]):
        for j in range(conf_mat.shape[1]):
            axe.text(j, i, plot_cm[i, j].astype(fmt),
                     ha="center", va="center",
                     color="white" if conf_mat[i, j] > thresh else "black")
    fig.tight_layout()
    return fig, axe

def confusion_matrix(true_label, pred_label, num_class):
    """
    Compute the confusion matrix
    Args:
        - true_label(list of int): list of true labels
        - pred_label(list of int): list of predicted labels

    Returns:
        - confusion matrix
    """
    cm = np.zeros((num_class, num_class))
    for i, j in zip(true_label, pred_label):
        cm[i, j] += 1
    return cm
```

```python
ellipse = pd.read_csv('ellipse.csv')
polygone = pd.read_csv('polygone.csv')
hybrid = pd.read_csv('hybrid.csv')
infer_ellipse = pd.read_csv('infer_regression_ellipse.csv')
infer_polygone = pd.read_csv('infer_regression_polygone.csv')
regression_ellipse = pd.read_csv('regression_ellipse.csv')
regression_polygone = pd.read_csv('regression_polygone.csv')
```

```python
ellipse["ellipse_comp"] = np.where(ellipse['ellipses'] == ellipse[' predicted ellipses'], True, False)
polygone["polygone_comp"] = np.where(polygone['polygone'] == polygone[' predicted polygones'], True, False)
hybrid["ellipse_comp"] = np.where(hybrid['ellipses'] == hybrid[' predicted ellipses'], True, False)
infer_ellipse["ellipse_comp"] = np.where(infer_ellipse['ellipses'] == infer_ellipse[' predicted ellipses'], True, False)
infer_polygone["polygone_comp"] = np.where(infer_polygone['polygone'] == infer_polygone[' predicted polygones'], True, False)
regression_ellipse["ellipse_comp"] = np.where(regression_ellipse['ellipses'] == regression_ellipse[' predicted ellipses'], True, False)
regression_polygone["polygone_comp"] = np.where(regression_polygone['polygone'] == regression_polygone[' predicted polygones'], True, False)
hybrid["polygone_comp"] = np.where(hybrid[' polygones'] == hybrid[' predicted polygones'], True, False)
```

```python
print(ellipse["ellipse_comp"].mean())
print(polygone["polygone_comp"].mean())
print(hybrid["ellipse_comp"].mean())
print(infer_ellipse["ellipse_comp"].mean())
print(infer_polygone["polygone_comp"].mean())
print(regression_ellipse["ellipse_comp"].mean())
print(regression_polygone["polygone_comp"].mean())
print(hybrid["polygone_comp"].mean())
```

```python
classes = [str(i) for i in range(6)]
cm = confusion_matrix(ellipse['ellipses'], ellipse[' predicted ellipses'], 6)
plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title="Ellipse confusion matrix",
                          cmap=plt.cm.Blues)
classes = [str(i) for i in range(6)]
cm = confusion_matrix(polygone['polygone'], polygone[' predicted polygones'], 6)
plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title="polygones confusion matrix",
                          cmap=plt.cm.Blues)
cm = confusion_matrix(hybrid['ellipses'], hybrid[' predicted ellipses'], 6)
plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title="Ellipse confusion matrix",
                          cmap=plt.cm.Blues)
cm = confusion_matrix(hybrid[' polygones'], hybrid[' predicted polygones'], 6)
plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title="polygones confusion matrix",
                          cmap=plt.cm.Blues)
```

```python
classes = [str(i) for i in range(6)]
regression_ellipse = regression_ellipse.clip(upper=5, lower=0)
regression_polygone = regression_polygone.clip(upper=5, lower=0)
cm = confusion_matrix(infer_ellipse['ellipses'], infer_ellipse[' predicted ellipses'], 6)
plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title="Inferred regression ellipse",
                          cmap=plt.cm.Blues)
classes = [str(i) for i in range(6)]
cm = confusion_matrix(infer_polygone['polygone'], infer_polygone[' predicted polygones'], 6)
plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title="Inferred regression polygones",
                          cmap=plt.cm.Blues)
cm = confusion_matrix(regression_ellipse['ellipses'], regression_ellipse[' predicted ellipses'], 6)
plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title="regression Ellipse",
                          cmap=plt.cm.Blues)
cm = confusion_matrix(regression_polygone['polygone'], regression_polygone[' predicted polygones'], 6)
plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title="regression polygones",
                          cmap=plt.cm.Blues)
```
